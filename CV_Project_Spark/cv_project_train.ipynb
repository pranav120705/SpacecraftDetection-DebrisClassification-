```python
from ultralytics import YOLO

# === CONFIGURATION ===
model_path = 'yolov8m.pt'                   # Pretrained model or checkpoint to start from
data_path = "/kaggle/input/spark-cv-prj11/sparkfull/data.yaml"                    # Path to dataset YAML file
epochs = 50                                 # Total number of epochs to train
save_period = 1                            # Save model every epoch
resume = True                              # Set True if resuming from checkpoint
checkpoint_path = '/kaggle/input/cvlast11/other/default/1/last.pt'  # Used only if resume=True

# === LOAD MODEL ===
if resume:
    model = YOLO(checkpoint_path)           # Load from checkpoint
else:
    model = YOLO(model_path)                # Start from pretrained

model.train(
        data=data_path,
        epochs=epochs,
        imgsz=640,
        batch=4,
        device=0,
        workers=2,
        save_period=save_period,
        resume=resume,
        project='runs/detect',
        name='custom_run',
        verbose=True
    )
```
